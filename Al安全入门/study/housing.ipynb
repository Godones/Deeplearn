```python
import os
import torch
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import torch.utils.data as Data
import torch.nn as nn
import torch.optim as optim
from torch.nn import init               
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("boston_housing_data.csv")
data.head()
```

```python
data.keys(),data.index
data = data.dropna()
data.head()
```

```python
y = data['MEDV']
x = pd.DataFrame(data,columns=['CRIM', 'INDUS', 'NOX',
                               'RM', 'AGE', 'DIS',
                               'TAX','PIRATIO', 'B', 'LSTAT'])
x.shape,y.shape
```

```python
features = torch.tensor(np.array(x),dtype=torch.float32)
labels = torch.tensor(y,dtype=torch.float32).view(len(y),1)
features.shape,labels.shape
```

```python
batch_size = 20
dataset = Data.TensorDataset(features,labels)
data_iter = Data.DataLoader(dataset,batch_size=batch_size,shuffle=True)
for X,y in data_iter:
    print(X.shape,y.shape)
    break
```

```python
class LinearNet(nn.Module):
    def __init__(self,n_feature):
        super(LinearNet,self).__init__()
        self.Linear = nn.Linear(n_feature,1)
    def forward(self,x):
        y = self.Linear(x)
        # print(y)
        return y
```

```python
net = LinearNet(10)
print(net)
for param in net.parameters():
    print(param)
```

```python
loss = nn.MSELoss()
# optimizer = optim.SGD(net.parameters(),lr=1e-6)
optimizer = optim.Adam(net.parameters(),lr=0.03)
print(optimizer,loss)
```

```python
num_epoch = 1000
loss_record = []
for epoch in range(num_epoch):
    for X,y in data_iter:
        out_put = net(X)
        l = loss(out_put,y)
        # print(l)
        optimizer.zero_grad()
        l.backward()
        optimizer.step()
    print('epoch %d, loss: %f' % (epoch, l.item()))
    loss_record.append(l.item())
```
